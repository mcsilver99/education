# Education Project: Regression

### Import libraries

```python
# Import pandas, numpy, and matplotlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# seaborn is a data visualization library built on matplotlib
import seaborn as sns

# set the plotting style
sns.set_style("whitegrid")

# Import stat model
import statsmodels.formula.api as smf
from statsmodels.formula.api import ols
from sklearn.metrics import *
```

Load clean data set

```python
df = pd.read_csv('https://raw.githubusercontent.com/mcsilver99/education/main/data/education_clean.csv')
```

### Modeling

#### Single input models

Fit and assess models predciting average ACT score from each of the input variables. You might try polynomial linear regression models, as appropriate. 

#### Median Income

##### Plot the regression line and the scatter plot

```python
plt.figure(figsize=(6,6))

sns.regplot(data=df, 
            x = 'median_income',
            y='average_act',
            color='blue',
            ci=False, 
            scatter_kws={'color':'black', 'edgecolors':'white', 'linewidths':1})

# Add axel labels
plt.xlabel('Median income ($)', fontsize=16)
plt.ylabel('Average ACT score', fontsize=16)

# Increase the fontsize of the tick labels
plt.tick_params(labelsize=14)

plt.show()
            
```

**Note:** There is a relationship between average ACT acore and median income. It appears this simple linear regression is providing a moderate fit. 

##### Fit the simple linear regression model

```python
model_median_income = smf.ols(formula='average_act ~ median_income', data=df).fit()
```

```python
print(model_median_income.summary())
```

**Note:** The intercept is 17.8. The p-values are small so we have statistically significant coefficients. The R^2 is relatively low.

#### Numerical assessment of fit accuracy 

##### Compute the R-squared

```python
model_median_income.rsquared
```

##### Compute the RMSE

```python
y_hat = model_median_income.predict()
```

```python
np.sqrt(mean_squared_error(df['average_act'], y_hat)).round(3)
```

##### Compute the mean absolute error

```python
mean_absolute_error(df['average_act'], y_hat)
```

**Note:** On average the mean error between the regression error is 1.7. This is relatively small error given the range of possible ACT scores.

#### Use a redisual plot for graphical assessment of model fit

```python
plt.figure(figsize=(6,6))

plt.plot(y_hat, model_median_income.resid, 'ko', mec='w')
plt.axhline(0, color='r', linestyle='dashed', lw=2)

plt.xlabel('Predict average ACT score', fontsize=16)
plt.ylabel('Residual average ACT score', fontsize=16)

plt.tick_params(labelsize=14)

plt.show()
```

**Note:** We can consider a quadratic model.

##### Try a quadratic model. Plot the regression curves and scatter plot.

```python
plt.figure(figsize=(6,6))

sns.regplot(data=df,
            x='median_income',
            y='average_act',
            color='blue',
            ci=False,
            scatter_kws={'color':'black', 'edgecolors':'white', 'linewidths':1})

sns.regplot(data=df,
            x='median_income',
            y='average_act',
            order = 2,
            color='orange',
            ci=False, 
            scatter=False)

# Add axis labels
plt.xlabel('Median income ($)', fontsize=16)
plt.ylabel('Average ACT score', fontsize=16)
```

**Note:** The quadratic model mighht provide a slighhtly better fit but we can't tell if it is significantly better. 

##### Fit a quadratic linear regression model

```python
model_median_income2 = smf.ols(
    formula='average_act ~ median_income + I(median_income**2)',
    data=df).fit()
```

```python
print(model_median_income2.summary())
```

**Note:** The R^2 is slightly highher than the previous model. The coefficients are statistically significant but it is not clear how much it is improving the model.

##### Assess the model significance

```python
model_median_income2.pvalues
```

```python
from statsmodels.stats.anova import anova_lm
anova_lm(model_median_income, model_median_income2)
```

**Note:** The p-value being quite small indicates the more complex model is statistcally different from the simple model.

##### Compare the quadratic model accuracy to the first order model

```python
mean_absolute_error(df['average_act'], model_median_income2.predict())
```

```python
mean_absolute_error(df['average_act'], model_median_income.predict())
```

**Note:** The error is smaller but not that much smaller. The quadratic model might only be slightly better so the simple linear regression model is probably sufficient. 

### Multiple Linear Regression

```python
model = smf.ols(
    formula='average_act ~ rate_unemployment + percent_college + percent_married + median_income + percent_lunch',
    data=df).fit()
```

```python
print(model.summary())
```

**Note:** Percent married and median income produce p-values that are not statistcally significant. There was correlations among the predictors so we may not have to include all of them in the model.  

#### Use a residual plot for graphical assessment of model fit

```python
y_hat = model.predict()
```

```python
plt.figure(figsize=(5,5))

plt.plot(y_hat, model.resid, 'ko', mec = 'w')
plt.axhline(0, color='r', linestyle='dashed', lw=2)

plt.xlabel('Predicted average ACT score', fontsize=16)
plt.ylabel('Residual average ACT score', fontsize=16)

plt.tick_params(labelsize=14)

plt.show()
```

**Note:** There is relatively no structure to the residual plot so transformations of the predictors are unlikely to help the model. 

##### Compute the mean absolute error

```python
mean_absolute_error(df['average_act'], model.predict())
```

**Summary:**
1. The multiple linear regression model with all socioeconmic predictors can predict ACT scores with an average error of roughly 1 point using these predictors
2. The residual plots suggests that there will be limited improvement from adding trasnformations of these predictors
3. Not all predictors have significant coefficients.
4. Adding predictors to the model may imrpove the performance. 

### Fit reduced model with significant predictors

```python
model_reduced = smf.ols(
    formula='average_act ~ rate_unemployment + percent_college + percent_lunch',
    data=df).fit()
```

```python
print(model_reduced.summary())
```

**Note:** This model exlcudes the statistcally insignificant predictors from the full model. We can see that all predictors have statistcally significance. 

```python
plt.figure(figsize=(5,5))

plt.plot(y_hat, model_reduced.resid, 'ko', mec = 'w')
plt.axhline(0, color='r', linestyle='dashed', lw=2)

plt.xlabel('Predicted average ACT score', fontsize=16)
plt.ylabel('Residual average ACT score', fontsize=16)

plt.tick_params(labelsize=14)

plt.show()
```

**Note:** No structure to the plot so we will not have to perform any transformations of the predictors. 

##### Compute the mean absolute error

```python
mean_absolute_error(df['average_act'], model_reduced.predict())
```

#### Compare the accuracy between the full and reduced models

```python
mae_full = mean_absolute_error(df['average_act'], model.predict())
mae_reduced = mean_absolute_error(df['average_act'], model_reduced.predict())

r1_full = model.rsquared
r2_reduced = model_reduced.rsquared

pd.DataFrame({'Mean Absolute Error': [mae_full, mae_reduced], 
              'R-squared': [r1_full, r2_reduced]}, 
             index = ['full model', 'reduced model']).round(4)
```

**Note** There is not much of a difference in accuracy between thhe full and reduced models. 

#### Use an ANOVA to test the significance of difference between models

```python
anova_lm(model_reduced, model)
```

**Summary:**
1. The model preducting the average ACT score from rate_unemployment, percent_college, and percent_lunch is not significantly different than the model with all five socioeconomic variables as predictors.
2. Adding predictors to the model may improve performance.
3. We have not yet determined the relative importance of each predictor in the model. 

### Scaling

##### Scale the predictor variables in the reduced model to have mean 0 and standard deviation 1 and add them to the data frame.

```python
predictor_variables = ['rate_unemployment', 'percent_college', 'percent_lunch']
```

```python
scaled_columns = [var + '_normalized' for var in predictor_variables]
print(scaled_columns)
```

**The standardization is making the predictors have mean 0 and SD 1.** 

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(df[predictor_variables])
```

```python
df[scaled_columns] = scaler.transform(df[predictor_variables])
```

##### Check the mean and standard deviation of the transformed data

```python
df[scaled_columns].agg(['mean', 'std']).round(3)
```

**Now we have standardized or normalized predictors.**

##### Fit the multiple linear regression model with the normalized predictors

```python
model_normalized = smf.ols(
    formula='average_act ~ rate_unemployment_normalized + percent_college_normalized + percent_lunch_normalized',
    data=df).fit()
```

```python
print(model_normalized.summary())
```

**Note:** The coefficient for perecent lunch is of much larger magnitude compared to the other predictors. 

#### Compare the accuracy between the original and normalized models

```python
mae_normalized = mean_absolute_error(df['average_act'], model_normalized.predict())
mae_reduced = mean_absolute_error(df['average_act'], model_reduced.predict())

r2_normalized = model_normalized.rsquared
r2_reduced = model_reduced.rsquared

pd.DataFrame({'Mean Absolute Error': [mae_normalized, mae_reduced], 
              'R-squared': [r2_normalized, r2_reduced]}, 
             index = ['normalized model', 'reduced model']).round(4)
```

**Note:** Normalizing the predictors in the model does not lead to an overall change to the linear regression model.

**Summary:**
1. The percent of students at a school receiving free or reduced price lunch was the strongest predictor in this analysis.
